```python
fig,axes = plt.subplots(nrows=2, ncols = 1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=3, ncols = 1, figsize=(10,4))

axes[0].plot(x,y)
axes[1].plot(y,x)
axes[2].plot(x,y)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
# Add a legend

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'X Cubed')

ax.legend(loc = 0)

# loc = 0, this allows for the best area for the legend to be chosen
# You can customze the number to your needs
```
